```python
t = var("t") 
var("L, C, t0, q0, I0") 
q = function("q")(t) 
assume(L>0);assume(C>0);assume(t>0) 
LC = L*diff(q,t,2)+q/C == 0
LCo = desolve(LC, q, ivar=t)
q_LC = desolve(LC, q, ivar=t, ics=[t0,q0,I0])
q_I=diff(q_LC, t, 1)
Upl=q_LC.substitute(t0=0, q0=0.5,
         I0=0, L=12e-4, C=1.5e-2);
Ipl=q_I.substitute(t0=1, q0=1e-4,
         I0=1, L=12e-2, C=1.5e-4);
I0=plot(Ipl, (0, 0.1),thickness=0.5,color="green")
U0=plot(Upl, (0, 0.1),thickness=0.5,color="brown")
show(I0+U0,figsize=2,fontsize=5)
```
